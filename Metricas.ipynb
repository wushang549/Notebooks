# Justifiacion

El problema es de clasificación binaria con clases desbalanceadas, donde la clase positiva (clientes que sí contratan el producto) es minoritaria. En este contexto, la exactitud puede resultar engañosa, ya que un modelo podría obtener valores altos simplemente prediciendo siempre la clase mayoritaria.

El F1-score combina precisión y recall en una sola métrica, penalizando tanto los falsos positivos como los falsos negativos. Esto es adecuado para el problema, ya que:

Es importante no perder clientes potenciales (recall).
También es importante no contactar clientes que no contratarán (precisión).

Por lo tanto, el F1-score representa un equilibrio adecuado entre ambos.
